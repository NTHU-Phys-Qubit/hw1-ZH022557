```python
import numpy as np
import matplotlib.pyplot as plt
from time import time 
```

# 1D Random walk

We will concider two kinds of 1D random talk.
* $dx$ is uniformly distributed between $[-1/2, +1/2]$.
* $dx=+1$ with probability 1/2 and $dx=-1$ with probability 1/2.

For both kinds of the 1D random walk, you should do the following:
* Plot a single trajectory.
* Plot the histogram of the end point for an ensemble of trajectories.

```python
t=time()
x,position=0,[]
for j in np.arange(1000):
        dx = np.random.random()-0.5
        x  = x + dx
        position.append(x)
final=[x]
for i in np.arange(2000):
    x = 0
    for j in np.arange(1000):
        dx = np.random.random()-0.5
        x  = x + dx
    final.append(x)
mu, sigma = 0, np.std(final, ddof = 1)
sx = np.linspace(-1*max(final),max(final),50)
sy = 1/(sigma * np.sqrt(2 * np.pi)) * np.exp( - (sx - mu)**2 / (2 * sigma**2) )
plt.figure(figsize=(30,10))
plt.subplot(1,3,1)
plt.plot(position)
plt.subplot(1,3,2)
plt.hist(final, bins=60)
plt.subplot(1,3,3)
plt.plot(sx, sy)
plt.ylim(0,max(sy)*1.1)
t=time()-t
print(t)
```
